Image to Test using StableDiffusion

```python
import time
import keras_cv
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
keras.mixed_precision.set_global_policy("mixed_float16")
```

```python
model = keras_cv.models.StableDiffusion(jit_compile=True)
```

```python
image = model.text_to_image("Photograph of siren eyes", batch_size=3)

def plot_image(images):
    plt.figure(figsize=(20,20))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[1])
        plt.axis("off")
        
plot_image(image)
```

```python
def plot_image1(images):
    plt.figure(figsize=(50,50))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[1])
        plt.axis("off")
        plt.savefig("siren_eyes.jpg")

plot_image1(image)
```
